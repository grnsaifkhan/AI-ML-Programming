1. Create a text file named emails.txt containing multiple email addresses,
some in uppercase and some repeated.

Write a Python program to,

*   clean the data by converting all emails to lowercase
*   removing duplicates
*   sorting them alphabetically
*   saving the cleaned list to a new file named unique_emails.txt

```python
with open("./sample_data/emails.txt","r") as file:
  emails = file.readlines()

  #cleaning process
  lowered_char_emails = set(map(lambda x: x.lower(), emails))
  #new line delete
  unique_lowered_char_emails = map(str.strip, lowered_char_emails)
  sorted_unique_lowered_char_emails = sorted(unique_lowered_char_emails)

  print(sorted_unique_lowered_char_emails)

with open("./sample_data/unique_emails.txt","w") as file:
  for email in sorted_unique_lowered_char_emails:
    file.write(f"{email}\n")
  # print(file.tell())
  file.truncate(file.tell()-1)
```

Create a text file named names.txt containing a list of names.\
Some names may have extra spaces, inconsistent capitalization (like “aLiCe” or “ bob ”), or appear more than once.\
Write a Python program to clean the data by:
1.  Removing extra spaces.
2.  Converting all names so that only the first letter is capitalized.
3.  Removing duplicate names.
4. Sorting the names alphabetically.\
Finally, save the cleaned list into a new file named clean_names.txt.

```python
with open("./sample_data/names.txt", "r") as file:
  name_list = file.readlines()

  #remove extra space
  names = set(map(lambda x: x.replace(" ","").capitalize().strip(), name_list))

  #sorted name
  names = sorted(names)
  print(names)


  with open("./sample_data/clean_names.txt", "w") as file:
    for name in names:
      file.write(f"{name}\n")
    # print(file.tell())
    file.truncate(file.tell()-1)
```

Write a Python program that creates a list of 5 numbers.\
Ask the user for an index number and print the value at that index.\
If the index is out of range, handle the exception and print "Invalid index!"

```python
list_value = [1,2,3,4,5]

try:
  input_index = int(input("Enter an index (0-4): "))
  print("Value at index", input_index, "is:", list_value[input_index])
except IndexError:
  print("Invalid index!")
```
